```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
RF_file =  "2025-1 고물실 data/20250617_204338_pixel_x_101_pixel_y_101_RF.csv"
No_RF_file = "2025-1 고물실 data/20250617_204014_pixel_x_101_pixel_y_101_no_RF.csv"
mirror_amp = -0.2
pixel_x = 101
pixel_y = 101
```

```python
df_rf = pd.read_csv(RF_file)
df_no_rf = pd.read_csv(No_RF_file)
```

```python
rf_array = np.array(df_rf.values)
no_rf_array = np.array(df_no_rf.values)
```

```python
delta_counts_array = no_rf_array - rf_array
```

```python
image = plt.imshow(no_rf_array,
                  extent=[mirror_amp, -mirror_amp,mirror_amp, -mirror_amp],origin="lower",
                  aspect='auto', cmap='coolwarm')

cbar = plt.colorbar(image,label="Photon Counts/s")
```

```python
voltage_x = np.linspace(mirror_amp, -mirror_amp, pixel_x)
voltage_y = np.linspace(mirror_amp, -mirror_amp, pixel_x)
```

```python
start_x = -0.15
end_x = -0.05
start_y = 0.01
end_y = 0.1

start_x_idx = np.where(voltage_x >= start_x)[0][0]
start_y_idx = np.where(voltage_y >= start_y)[0][0]
end_x_idx = np.where(voltage_x <= end_x)[0][-1]
end_y_idx = np.where(voltage_y <= end_y)[0][-1]

```

```python









fig, axs = plt.subplots(1,3,figsize=(40,10))





image0 = axs[0].imshow(no_rf_array[start_y_idx:end_y_idx][start_x_idx:end_x_idx],
                  extent=[start_x, end_x,start_y, end_y],origin="lower",
                  aspect='auto', cmap='coolwarm')

image1 = axs[1].imshow(rf_array[start_y_idx:end_y_idx][start_x_idx:end_x_idx],
                  extent=[start_x, end_x,start_y, end_y],origin="lower",
                  aspect='auto', cmap='coolwarm')

image2 = axs[2].imshow(delta_counts_array[start_y_idx:end_y_idx][start_x_idx:end_x_idx],
                  extent=[start_x, end_x,start_y, end_y],origin="lower",
                  aspect='auto', cmap='coolwarm', vmin=0)



axs[0].set_title("No RF")
axs[1].set_title("RF")
axs[2].set_title("Delta Counts")

cbar0 = plt.colorbar(image0,label="Photon Counts/s")
cbar1 = plt.colorbar(image1,label="Photon Counts/s")
cbar2 = plt.colorbar(image2,label="Photon Counts/s")
```
